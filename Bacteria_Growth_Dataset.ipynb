```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv('Intern/Bacteria_Growth_Dataset.csv')  # Change this if needed

# Display basic dataset information
df.head()
```

```python
df.info()
```

```python
# 1️⃣ Plot Bacterial Growth Over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Time (hrs)"], df["Bacterial Growth (CFU/mL)"], marker='o', linestyle='-')
plt.xlabel("Time (hrs)")
plt.ylabel("Bacterial Growth (CFU/mL)")
plt.title("Bacterial Growth Over Time")
plt.grid()
plt.show()
```

```python
# 2️⃣ Analyze Effect of Temperature & pH on Growth
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Temperature (°C)"], y=df["Bacterial Growth (CFU/mL)"], hue=df["pH"], palette="coolwarm")
plt.title("Effect of Temperature & pH on Bacterial Growth")
plt.xlabel("Temperature (°C)")
plt.ylabel("Bacterial Growth (CFU/mL)")
plt.show()
```

```python
# 3️⃣ Compare Growth Under Different Antibiotic Treatments
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Antibiotic (mg/L)"], y=df["Bacterial Growth (CFU/mL)"])
plt.xlabel("Antibiotic Concentration (mg/L)")
plt.ylabel("Bacterial Growth (CFU/mL)")
plt.title("Effect of Antibiotics on Bacterial Growth")
plt.show()
```

```python
# 4️⃣ Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# 5️⃣ Answering Key Research Questions

# ✅ Best Growth Conditions (Temperature & pH)
best_growth = df.loc[df["Bacterial Growth (CFU/mL)"].idxmax()]
best_temp, best_pH = best_growth["Temperature (°C)"], best_growth["pH"]
```

```python
# ✅ Most Effective Antibiotic Concentration for Growth Inhibition
lowest_growth = df.loc[df["Bacterial Growth (CFU/mL)"].idxmin()]
best_antibiotic = lowest_growth["Antibiotic (mg/L)"]

print(f"Bacterium X grows the fastest at {best_temp}°C and pH {best_pH}")
print(f"The most effective antibiotic concentration for inhibition is {best_antibiotic} mg/L")
```

```python
# 6️⃣ Line Plot for Growth Trends
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["Time (hrs)"], y=df["Bacterial Growth (CFU/mL)"], marker='o', linestyle='-')
plt.xlabel("Time (hrs)")
plt.ylabel("Bacterial Growth (CFU/mL)")
plt.title("Bacterial Growth Trends Over Time")
plt.grid()
plt.show()
```

```python
# 7️⃣ Scatter Plot for Antibiotic Effectiveness
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Antibiotic (mg/L)"], y=df["Bacterial Growth (CFU/mL)"], alpha=0.7)
plt.xlabel("Antibiotic Concentration (mg/L)")
plt.ylabel("Bacterial Growth (CFU/mL)")
plt.title("Effect of Antibiotic Concentration on Bacterial Growth")
plt.show()
```
